## Regression Model Evaluation

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
"""
CRIM - per capita crime rate by town
ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
INDUS - proportion of non-retail business acres per town.
CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX - nitric oxides concentration (parts per 10 million)
RM - average number of rooms per dwelling
AGE - proportion of owner-occupied units built prior to 1940
DIS - weighted distances to five Boston employment centres
RAD - index of accessibility to radial highways
TAX - full-value property-tax rate per $10,000
PTRATIO - pupil-teacher ratio by town
B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
LSTAT - % lower status of the population
MEDV - Median value of owner-occupied homes in $1000's"""
```

```python
data
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
# Your code here :
from sklearn.model_selection import train_test_split
X = data[['CRIM', 'AGE', 'RAD', 'LSTAT']]
y = data[['MEDV']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python
# Your code here :
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import r2_score

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print("R-squared (Training set):", r2_train)
print("R-squared (Testing set):", r2_test)
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import mean_squared_error

mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print("Mean Squared Error (Training set):", mse_train)
print("Mean Squared Error (Testing set):", mse_test)
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import mean_absolute_error

mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

print("Mean Absolute Error (Training set):", mae_train)
print("Mean Absolute Error (Testing set):", mae_test)
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
# Your code here :
df['species'] = data.target
X = df[data.feature_names]  # now X is a DataFrame
y = df['species']           # target column

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
# Your code here :
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=200, solver='lbfgs', multi_class='auto')
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)

# Predict on test set
y_pred_test = model.predict(X_test)

print(len(y_train), len(y_pred_train))
print(len(y_test), len(y_pred_test))
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import accuracy_score
# Training set accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)

# Test set accuracy
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy:.3f}")
print(f"Testing Accuracy: {test_accuracy:.3f}")
```

## 9. Calculate and print the balanced accuracy score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import balanced_accuracy_score
# Training set
balanced_train = balanced_accuracy_score(y_train, y_pred_train)

# Test set
balanced_test = balanced_accuracy_score(y_test, y_pred_test)

print(f"Balanced Accuracy (Train): {balanced_train:.3f}")
print(f"Balanced Accuracy (Test): {balanced_test:.3f}")
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import precision_score
# Training set
precision_train = precision_score(y_train, y_pred_train, average='weighted')

# Test set
precision_test = precision_score(y_test, y_pred_test, average='weighted')

print(f"Precision (Train): {precision_train:.3f}")
print(f"Precision (Test): {precision_test:.3f}")
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import recall_score
recall_train = recall_score(y_train, y_pred_train, average='weighted')

recall_test = recall_score(y_test, y_pred_test, average='weighted')

print(f"Recall (Train): {recall_train:.3f}")
print(f"Recall (Test): {recall_test:.3f}")
```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import f1_score

f1_train = f1_score(y_train, y_pred_train, average='weighted')

f1_test = f1_score(y_test, y_pred_test, average='weighted')

print(f"F1 Score (Train): {f1_train:.3f}")
print(f"F1 Score (Test): {f1_test:.3f}")
```

## 13. Generate confusion matrices for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Training set confusion matrix
cm_train = confusion_matrix(y_train, y_pred_train)

# Test set confusion matrix
cm_test = confusion_matrix(y_test, y_pred_test)

# Training set
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train,
                                    display_labels=data.target_names)
disp_train.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix - Training Set")
plt.show()

# Test set
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test,
                                   display_labels=data.target_names)
disp_test.plot(cmap='Greens', values_format='d')
plt.title("Confusion Matrix - Test Set")
plt.show()
```
